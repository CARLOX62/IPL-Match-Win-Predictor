```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Punjab Kings',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Lucknow Super Giants',
    'Gujarat Titans'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df
```

```python
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
delivery_df = match_df.merge(delivery, on = 'match_id')
```

```python
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df
```

```python
delivery_df.groupby('match_id')['total_runs_y'].transform('cumsum')
```

# Current Score

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].transform('cumsum')
```

```python
delivery_df
```

# Runs Left

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

# Ball Left

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

# Getting Wickets

```python
# Player dismissed tell if batsmen was out or not while playing that ball
# filling null value with 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")

# If coloumn value is not zero mark it 1
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")

# Changing datatype of column name
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')

# Finding wickets using cumulative sum functionj
wickets = delivery_df.groupby('match_id')['player_dismissed'].transform('cumsum').values

# calculating wickets left by simple maths operation
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df.tail()
```

# CRR (Current Run Rate)

```python
# crr = runs/overs

delivery_df['crr'] = round((delivery_df['current_score']*6)/(120 - delivery_df['balls_left']), 2)
```

```python
delivery_df
```

```python
delivery_df['crr'].isnull().sum()
```

```python
delivery_df['crr'].fillna(delivery_df['crr'].mean(), inplace = True)
```

```python
delivery_df = delivery_df[delivery_df['balls_left'] != 0]
```

# RRR (Required Run Rate)

```python
delivery_df['rrr'] = round((delivery_df['runs_left']*6)/delivery_df['balls_left'], 2)
```

```python
delivery_df
```

# Finding Winner

```python
# Function to mark winner team 0 or 1 based on winner column
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis = 1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wickets', 'total_runs_x', 'crr', 'rrr', 'result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

# Dropping Null values

```python
final_df.dropna(inplace = True)
```

# Remove rows where balls_left == 0 to remove -inf and inf from rrr column

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df
```

# CRR have a infinite value 

```python
final_df.to_csv('final_df.csv', index = False)
```

```python
np.isinf(final_df['crr']).sum()
```

```python
final_df['crr'] = final_df['crr'].replace([np.inf, -np.inf], np.nan)
final_df['crr'] = final_df['crr'].fillna(final_df['crr'].median())
```

# Creating the model

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state = 100)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf1', OneHotEncoder(sparse_output = False, drop = 'first'), ['batting_team', 'bowling_team', 'city'])
], remainder = 'passthrough' )
```

```python
from sklearn.linear_model import LogisticRegression
# from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
# Creating pipeline (building model) 
pipe = Pipeline([
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
X_train.describe()
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
